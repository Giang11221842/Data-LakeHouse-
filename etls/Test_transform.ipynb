```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, lit, year, month,
    to_timestamp, unix_timestamp
)

spark = SparkSession.builder \
    .appName("Taxi Silver Layer") \
    .getOrCreate()


```

```python
df_yellow = spark.read.parquet("/home/njagz/Documents/FInal_Projects/data/input/yellow/year=2022/month=01/data.parquet")
df_green = spark.read.parquet("/home/njagz/Documents/FInal_Projects/data/input/green/year=2022/month=01/data.parquet")
```

```python
# Transformations for Yellow Taxi Data
yellow_transformed = df_yellow.withColumnRenamed("tpep_pickup_datetime", "pickup_datetime") \
    .withColumnRenamed("tpep_dropoff_datetime", "dropoff_datetime") \
    .withColumn("service_type", lit("yellow")) \
    .withColumn("RatecodeID", col("RatecodeID").cast("integer")) \
    .withColumn("passenger_count", col("passenger_count").cast("integer"))  \
    .withColumn("trip_type", lit(0))  # Assuming trip_type is 0 for yellow taxis

# Transformations for Green Taxi Data
green_transformed = df_green.withColumnRenamed("lpep_pickup_datetime", "pickup_datetime") \
    .withColumnRenamed("lpep_dropoff_datetime", "dropoff_datetime") \
    .withColumn("service_type", lit("green")) \
    .drop("ehail_fee") \
    .withColumn("RatecodeID", col("RatecodeID").cast("integer")) \
    .withColumn("passenger_count", col("passenger_count").cast("integer")) \
    .withColumn("trip_type", col("trip_type").cast("integer")) \
    .withColumn("payment_type", col("payment_type").cast("integer")) \
    .withColumn("airport_fee", lit(0.00))  # Assuming airpot_fee is 0 for green taxis
```

```python
yellow_transformed = yellow_transformed.fillna({
    "store_and_fwd_flag": "N",
    "RatecodeID": -1,
    "passenger_count": 1,
    "congestion_surcharge": 0.0,
    "airport_fee": 0.0
})

green_transformed = green_transformed.fillna({
    "store_and_fwd_flag": "N",
    "RatecodeID": 0,
    "passenger_count": 0,
    "congestion_surcharge": 0.0,
    "payment_type": 0,
    "trip_type": -1
})
```

```python
def check_nulls(df):
    from pyspark.sql.functions import col, sum, when
    
    return df.select([
        sum(when(col(c).isNull(), 1).otherwise(0)).alias(c)
        for c in df.columns
    ])
```

```python
check_nulls(yellow_transformed).show()
check_nulls(green_transformed).show()
```
